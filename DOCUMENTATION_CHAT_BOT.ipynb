```python
%%writefile app.py
import streamlit as st
from langchain.chains import RetrievalQA
import pickle

# Load the QA chain from disk
with open("qa_chain.pkl", "rb") as f:
    qa_chain = pickle.load(f)

st.set_page_config(page_title="AI/ML Intern PDF Q&A", page_icon="🤖")
st.title(" Ask Questions from Document")

st.subheader("Ask anything below ")
query = st.text_input(" Your Question")
if st.button(" Submit"):
    if query:
        result = qa_chain.run(query)
        st.success(f" Answer: {result}")
    else:
        st.warning(" Please enter a question.")
```
